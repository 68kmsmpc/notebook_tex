```python
from qiskit import QuantumCircuit
from qiskit import circuit
```

```python

qc = QuantumCircuit(2, 2)
Ansatz = circuit.Gate(name="Ansatz", num_qubits=2, params = [2] ,label=None, duration=None )
qc.h(0)
qc.cx(0, 1)
qc.measure([0, 1], [0, 1])
qc.draw('mpl')
```

```python
from qiskit import QuantumCircuit
from qiskit.circuit import Gate

qc = QuantumCircuit(4, 1)
qc.append(Gate(name='Trial State \n (Parameterized Quantum Circuit)', num_qubits=4, params=[]), [0,1,2,3])
qc.append(Gate(name='basis\nchange', num_qubits=4, params=[]), [0,1,2,3])
qc.measure(0,0)
qc.measure(1,0)
qc.measure(2,0)
qc.measure(3,0)

qc.draw('mpl',style='iqp')
#,filename='circuit.png'
```
