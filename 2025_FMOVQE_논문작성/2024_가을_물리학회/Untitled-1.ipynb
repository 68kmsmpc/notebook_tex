```python
import numpy as np
import matplotlib.pyplot as plt
x= [0.78 ,0.94, 1]
E_VQE = [-1538.048824,-1538.10099,-1538.1604305776145] 
E_exact = [-1538.2228110340484,-1538.2675014744773 ,-1538.3341125776665] 
plt.figure(figsize=(6,4))
plt.plot(x,E_VQE, color='blue',label='E_FMO/VQE')
plt.scatter(x,E_VQE,color='blue',marker='x',s=80)
plt.plot(x,E_exact,color='red',ls='--',label='E_NumpyminimumEigensolver')
plt.scatter(x,E_exact,color='red',marker='x',s=80)
plt.xlabel('oxidation number')
plt.ylabel('Ground-State Energy(a.u.)')
plt.title('Energy for Three oxidation number')
plt.legend()
plt.ylim([-1538.5, -1538.0]) 
plt.grid()
plt.savefig('temp.png',dpi=1200)
```
